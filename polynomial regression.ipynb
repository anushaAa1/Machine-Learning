```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv('C:/Users/Lenovo/Downloads/Salary_Data.csv');
data.head()

```

```python
x=data['YearsExperience']
y=data['Salary']
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
p=np.polyfit(x,y,2)
f=np.poly1d(p)
print(f)
x_new = np.arange(min(x),max(x))
y_model = f(x_new)
plt.scatter(x,y)
plt.plot(x_new,y_model,'red')
plt.scatter(x_new,y_model)
```

```python
import numpy as np

x = data['YearsExperience']
y = data['Salary']
n = 2

A = np.zeros((n+1, n+1))
b = np.zeros((n+1,))

for i in range(n+1):
    print('row', i)
    for j in range(n+1):
        A[i, j] = np.sum(x**(i+j))
        print('row A element', f'sum(x**{i+j})')
        b[i] = np.sum(x**i * y)
        print('column b element', f'sum(x**{i+j})')

x = np.linalg.solve(A, b)
print(x)
```
